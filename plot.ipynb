```python
df = pd.read_csv('data.csv')
plt.ylabel('Time taken (s)')
plt.plot(df['Degree'], df["Karatsuba2"], label="Karatsuba 2")
plt.plot(df['Degree'], df['Schoolbook'], label="Schoolbook")
plt.plot(df['Degree'], df["Karatsuba32"], label="Karatsuba 32")
plt.plot(df['Degree'], df["Karatsuba"], label="Karatsuba")
plt.xlabel('Number of digits')
plt.xscale('log', base=10)
plt.yscale('log')
plt.title('Performance of the Karatsuba multiplication algorithm with different fallback thresholds')
plt.legend()
plt.grid(True)
plt.show()
```
